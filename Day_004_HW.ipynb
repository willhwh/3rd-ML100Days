#### 資料的 row 數以及 column 數

```python
print("row數:",len(app_train.index))
print("column數:",len(app_train.columns))
print(app_train.shape)
print(app_train.size)
```

```python
app_train.axes
```

#### 列出所有欄位

```python
ColumnName=[i for i in app_train.columns]
print(ColumnName)
```

```python
app_train.columns.tolist()
```

```python
app_train.index.tolist()
```

#### 截取部分資料

```python
app_train.loc[0:7]
#row0~row7
#[rows,columns]
#[0:7,:]
```

```python
app_train.iloc[[0,3,9],[3,7]]
#row0,3,9 and column3, 7
```

```python
app_train["TARGET"]==1
```

```python
app_train[app_train["TARGET"]==1]
#rows with TARGET==1 
```

```python
app_train[["TARGET","AMT_INCOME_TOTAL"]]
#all rows and column TARGET and AMT_INCOME_TOTAL
```

```python
app_train.iloc[:,0:3]
```

```python
app_train.iloc[0:3,3:6]
```
