### 載入 相關套件 並 執行 kmean 實驗 ...

- 實驗測試不同的群數
- 實驗測試不同的初始值
- 呈現結果

```python
from sklearn.cluster import KMeans
import numpy as np
```

```python
#實驗測試不同的群數
estimators = [('k_means_8', KMeans(n_clusters=8)),
              ('k_means_3', KMeans(n_clusters=3)),
#實驗測試不同的初始值            
              ('init_1',KMeans(n_clusters=3,init="random")),
             ('init_2',KMeans(n_clusters=3,init="random"))]
```

```python
fignum=1
title=["8 clusters","3 clusters","init_1","init_2"]
for name, est in estimators:
    est.fit(X)
    est.labels_
    est.cluster_centers_
    print(name,"\n",est.labels_,"\n",est.cluster_centers_,"\n")
```
